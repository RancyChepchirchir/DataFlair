```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
import numpy as np
# read csv
df = pd.read_csv("diabetes.csv")
X = df.iloc[:,0:8].values #tap du lieu train,test
y = df.iloc[:,8].values #validation

X_train,X_test, y_train, y_test = train_test_split(X,y)
print("X_train {}".format(X_train))
print("X_test {}".format(X_test))
print("y_train {}".format(y_train))
print("y_test {}".format(y_test))
clf = GaussianNB() # classifer
clf.fit(X_train,y_train)
print(clf)
y_predict = clf.predict(X_test)
print("predict:{}".format(y_predict))

#  count number predict accuracy (chinh xac)
predict_accuracy = [np.round(value) for value in y_predict]
accuracy = accuracy_score(y_test, predict_accuracy)
print("accuracy:%.2f" %(accuracy*100))
df.head()
```
