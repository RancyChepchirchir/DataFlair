```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
# random n and k cluster
n=200
k=2
x = -3 * np.random.rand(200,2)
x0 = 1+ 3*np.random.rand(100,2)
x[100:200,:] = x0
plt.scatter(x[:,0],x[:,1],s=25,color='r')
plt.grid()

# find centroids of 2 clusters
Kmean = KMeans(n_clusters=2)
Kmean.fit(x)
# print(Kmean.fit(x))
Kmean.cluster_centers_
print(Kmean.cluster_centers_)
plt.scatter(-1.44,-1.412,s=100,color='b')
plt.scatter(2.4282,2.3673, s=100, color='g')
#
Kmean.labels_
#
sample_text = np.array([4.0,5.0])
second_text = sample_text.reshape(1,-1)
print(sample_text)
print(second_text)
Kmean.predict(second_text)
```
