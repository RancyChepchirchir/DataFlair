```python
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
            AutoMinorLocator)
# each major's step = 20
majorLocator = MultipleLocator(20)
# major's format is digit
majorFormatter = FormatStrFormatter('%d')
# minor's step = 5
minorLocator = MultipleLocator(5)

# create sample values
t = np.arange(0.0, 100.0, 0.1)
s = np.sin(0.1 * np.pi * t) * np.exp(-t * 0.01)

# plotting
fig, ax = plt.subplots()
ax.plot(t,s,'r',linewidth='2')
# major setup
ax.xaxis.set_major_locator(majorLocator)
ax.xaxis.set_major_formatter(majorFormatter)

# for the minor ticks, use no labels; default NullFormatter
ax.xaxis.set_minor_locator(minorLocator)

plt.show()
```

```python
# subplot 1: with indexing
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(10, 3))

# 131 is nrows=1, ncols=3, pos=1
plt.subplot(141)
plt.bar(names, values)

# 132 is nrows=1, ncols=3, pos=2
plt.subplot(142)
plt.scatter(names, values)

# 133 is nrows=1, ncols=3, pos=3
plt.subplot(143)
plt.plot(names, values)

#
plt.subplot(144)
plt.scatter(names,values)

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
# subplots 2: with 1D array (nrows=1) or 2D array (nrows > 1)
names = ['A', 'B', 'C']
values = [1, 11, 100]

fig, axes = plt.subplots( nrows=3, ncols=1, figsize=(3,9))
axes[0].scatter(names, values)
axes[1].bar(names, values)
axes[2].plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
# subplots 3: exploding into tuptle of axes
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(9, 3))

ax1.bar(names, values)

ax2.scatter(names, values)

ax3.plot(names, values)

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
list_style = plt.style.available
print("Count: {}, list ={}".format(len(list_style),list_style))
```

```python
names =['A','B','C']
values=[1,10,100]

for i in range(len(list_style)):
    plt.style.use(list_style[i])
#     plt.style.use("bmh")
    plt.figure(figsize=(5,5))
    plt.bar(names, values)
    plt.title(list_style[i])
    plt.show()
```
